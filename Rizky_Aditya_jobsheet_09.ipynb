#Penugasan 03 - Eksperimen dengan Modul math

```python
import math

def hitung_luas_lingkaran(jari_jari):
    luas = math.pi * jari_jari**2
    return luas

r1 = 7
r2 = 10

print(f"Luas lingkaran dengan jari-jari {r1} adalah {hitung_luas_lingkaran(r1):.2f}")
print(f"Luas lingkaran dengan jari-jari {r2} adalah {hitung_luas_lingkaran(r2):.2f}")
```

#Penugasan 06 - Modul Buatan Sendiri

```python
%%writefile konversi.py

def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
%%writefile main_konversi.py

import konversi

celcius_temp = 25
fahrenheit_temp = 77

fahrenheit_converted = konversi.celcius_ke_fahrenheit(celcius_temp)
print(f"{celcius_temp} Celsius = {fahrenheit_converted:.2f} Fahrenheit")

celcius_converted = konversi.fahrenheit_ke_celcius(fahrenheit_temp)
print(f"{fahrenheit_temp} Fahrenheit = {celcius_converted:.2f} Celsius")
```

#Penugasan 08 - Studi Kasus Mini

```python
%%writefile laporan.py
def tampilkan_laporan(jumlah_soal, skor):
    print("\n=== Laporan Hasil Kuis ===")
    print(f"Jumlah Soal : {jumlah_soal}")
    print(f"Skor Akhir  : {skor}/{jumlah_soal}")
    if skor == jumlah_soal:
        print("Nice! Anda menjawab semua dengan benar.")
    elif skor >= jumlah_soal / 2:
        print("Cool! Anda dapat menjaawab beberapa dengan benar")
    else:
        print("Too bad, coba lagi!")
```

```python
%%writefile kuis.py
import random
import laporan

jumlah_soal = 5
skor = 0

print("=== Kuis Penjumlahan Sederhana ===")

for i in range(jumlah_soal):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    print(f"Soal {i+1}: {a} + {b} = ", end="")
    jawaban = int(input())

    if jawaban == a + b:
        print("Benar!\n")
        skor += 1
    else:
        print(f"Salah! Jawaban yang benar adalah {a + b}\n")

laporan.tampilkan_laporan(jumlah_soal, skor)
```
