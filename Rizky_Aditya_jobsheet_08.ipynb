#Praktikum 04 – Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

#Praktikum 07 – Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

#Penugasan 10 - Fungsi Decorator

```python
def logger(fungsi):
  def wrapper():
    print("Memulai Fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan.")
  return wrapper

@logger
def halo():
  print("Halo, Welcome boys!")

halo()
```

#Penugasan 11 - Studi Kasus Nyata – Program Nilai

```python
def input_nilai():
    """
    Meminta pengguna untuk memasukkan nilai-nilai (misalnya tugas, ujian)
    dan mengembalikannya dalam bentuk list of floats.
    """
    nilai_list = []
    print("Masukkan nilai-nilai (ketik 'selesai' untuk mengakhiri):")
    while True:
        try:
            input_user = input("> ")
            if input_user.lower() == 'selesai':
                break
            nilai = float(input_user)
            nilai_list.append(nilai)
        except ValueError:
            print("Input tidak valid. Harap masukkan angka atau 'selesai'.")
    return nilai_list


def hitung_rata_rata(nilai_list):
    """
    Menghitung rata-rata dari list nilai.
    """
    if not nilai_list:
        return 0.0
    return sum(nilai_list) / len(nilai_list)


def status_kelulusan(rata_rata, batas_lulus=60):
    """
    Menentukan status kelulusan berdasarkan rata-rata nilai.
    Lulus jika rata-rata >= batas_lulus.
    """
    if rata_rata >= batas_lulus:
        return "LULUS"
    else:
        return "TIDAK LULUS"


nilai_akhir_berbobot = lambda tugas, ujian: (0.40 * tugas) + (0.60 * ujian)



def main():
    print("--- Program Perhitungan Nilai Mahasiswa ---")

    # a
    while True:
        print("\nMasukkan 2 nilai untuk perhitungan (Tugas dan Ujian):")
        nilai_input = input_nilai()
        if len(nilai_input) == 2:
            break
        elif len(nilai_input) > 0:
            print("Harap masukkan TEPAT DUA nilai (Tugas dan Ujian) untuk menjalankan semua fitur.")
        else:
            print("Input nilai tidak boleh kosong.")

    nilai_tugas = nilai_input[0]
    nilai_ujian = nilai_input[1]

    # b
    nilai_akhir = nilai_akhir_berbobot(nilai_tugas, nilai_ujian)

    # c
    rata_rata_standar = hitung_rata_rata(nilai_input)

    # d
    status = status_kelulusan(nilai_akhir)

    print("\n--- Hasil Perhitungan ---")
    print(f"Nilai Tugas: {nilai_tugas}")
    print(f"Nilai Ujian: {nilai_ujian}")
    print(f"Rata-rata Nilai Standar: {rata_rata_standar:.2f}")
    print(f"**Nilai Akhir Berbobot (40% Tugas, 60% Ujian): {nilai_akhir:.2f}**")
    print(f"**Status Kelulusan (Batas Lulus >= 60): {status}**")
    print("---------------------------------------")

if __name__ == "__main__":
    main()
```
